```python
data[["pclass","survived"]].groupby(["pclass"]).mean()
```

```python
sns.catplot(x="sex",hue="survived",kind="count",data=data)
plt.xlabel("Sex")
plt.ylabel("Count")
plt.title("Survived")
```

```python
sns.catplot(x="sex",y="survived",hue="pclass",kind="bar",data=data)
plt.xlabel("Sex")
plt.ylabel("Survived")
plt.title("Survival Rate by Gender and Passenger Class")
```

```python
sns.catplot(x="sex",hue="pclass",kind="count",data=data)
plt.xlabel("Sex")
plt.ylabel("Count")
plt.title("Count of Passengers by Gender and Class")
```

```python
data.isnull()
```

```python
sns.catplot(x="embarked",hue="survived",kind="count",col="pclass",data=data)
plt.xlabel("Embarked")
plt.ylabel("Count")
plt.title("Passenger Count by Embarkation, Survival, and Class")
```

```python
sns.heatmap(data.isnull(),cbar="False",cmap="magma")
```

```python
sns.catplot(x="sex",y="survived",hue="pclass",kind="point",data=data)
plt.xlabel("Sex")
plt.ylabel("Survival Rate")
plt.title("Survival Rate by Gender and Passenger Class")
```

```python
sns.catplot(x="sex",y="survived",hue="pclass",kind="box",data=data)
plt.xlabel("Sex")
plt.ylabel("Survival Rate")
plt.title("Survival Rate by Gender and Passenger Class")
```

```python
sns.catplot(x="sex",y="survived",hue="pclass",kind="violin",data=data)
plt.xlabel("Sex")
plt.ylabel("Survival Rate")
plt.title("Survival Rate by Gender and Passenger Class")
```

```python
sns.catplot(x="sex",y="survived",hue="pclass",kind="strip",data=data)
plt.xlabel("Sex")
plt.ylabel("Survival Rate")
plt.title("Survival Rate by Gender and Passenger Class")
```

```python
sns.violinplot(x="sex",y="survived",hue="pclass",data=data)
plt.xlabel("Sex")
plt.ylabel("Survival Rate")
plt.title("Survival Rate by Gender and Passenger Class")
```
